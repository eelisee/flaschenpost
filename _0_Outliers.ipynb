```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df =pd.read_parquet(r"C:\Users\simon\OneDrive\Desktop\Datathon25\merged_df.parquet")
df
```

```python
df_filtered = df.drop(columns=['web_order_id','driver_id','customer_id_x','customer_id_y','warehouse_id','trip_id','total_weight','experience'])
df_filtered
```

```python
df_filtered = df_filtered[df_filtered['floor'] <= 50]
df_filtered = df_filtered[df_filtered['service_time_in_minutes'] <= 60]
df_filtered = df_filtered[df_filtered['article_count'] <= 230]

df_filtered
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_filtered)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df_filtered['article_count'])
plt.title('Boxplot of Article Count')
plt.show()


plt.figure(figsize=(10, 6))
sns.boxplot(x=df_filtered['floor'])
plt.title('Boxplot of Article Count')
plt.show()
```
