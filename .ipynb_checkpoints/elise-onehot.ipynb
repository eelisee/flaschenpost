```python
import pandas as pd
from IPython.display import display

# Load the parquet files
driver_order_mapping = pd.read_parquet('Flaschenpost_Datathon_2025_Dataset 1/masked_driver_order_mapping.parquet')
order_articles = pd.read_parquet('Flaschenpost_Datathon_2025_Dataset 1/masked_order_articles.parquet')
orders = pd.read_parquet('Flaschenpost_Datathon_2025_Dataset 1/masked_orders.parquet')
service_times = pd.read_parquet('Flaschenpost_Datathon_2025_Dataset 1/masked_service_times.parquet')

# Display the first few rows of each dataframe to verify the data
display(driver_order_mapping.head())
display(order_articles.head())
display(orders.head())
display(service_times.head())
```

```python
# Merge all dataframes on 'web_order_id'
merged_df = driver_order_mapping.merge(order_articles, on='web_order_id') \
                                .merge(orders, on='web_order_id') \
                                .merge(service_times, on='web_order_id')

# Display the first few rows of the merged dataframe to verify
display(merged_df.head())

# Print the number of rows in the merged dataframe
print(f"Number of rows in the merged dataframe: {len(merged_df)}")
```

```python
# Drop columns that end with '_y'
merged_df = merged_df.loc[:, ~merged_df.columns.str.endswith('_y')]

# Rename columns that end with '_x' to remove the suffix
merged_df.columns = merged_df.columns.str.replace('_x', '')

# Display the first few rows of the cleaned dataframe to verify
display(merged_df.head())
```

```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer

# Step 1: Group by web_order_id and aggregate columns
# For article_id, we aggregate into a list
aggregated = merged_df.groupby('web_order_id').agg({
    'article_id': lambda x: list(x),  # collect all article_ids for each order
    'box_id': pd.Series.nunique,      # count unique boxes per order
    'article_weight_in_g': 'sum',     # sum the weights per order
    # For remaining columns, take the first occurrence
    'driver_id': 'first',
    'warehouse_id': 'first',
    'order_time': 'first',
    'has_elevator': 'first',
    'floor': 'first',
    'is_pre_order': 'first',
    'is_business': 'first',
    'customer_id': 'first',
    'service_time_start': 'first',
    'service_time_end': 'first',
    'service_time_in_minutes': 'first',
    'order_datetime': 'first',
    'trip_id': 'first'
}).reset_index()

# Rename the aggregated columns as needed
aggregated.rename(columns={
    'box_id': 'unique_box_count',
    'article_weight_in_g': 'total_weight_in_g'
}, inplace=True)

# Step 2: One-hot encode the aggregated article_id lists using MultiLabelBinarizer
mlb = MultiLabelBinarizer()
one_hot_encoded = pd.DataFrame(mlb.fit_transform(aggregated['article_id']),
                               columns=[f"article_{cls}" for cls in mlb.classes_],
                               index=aggregated.index)

# Step 3: Concatenate the one-hot encoded columns with the aggregated dataframe
final_df = pd.concat([aggregated.drop(columns=['article_id']), one_hot_encoded], axis=1)

# Display the first few rows of the final dataframe
display(final_df.head())
```

```python
# One-hot encode the article_id column and attach the web_order_id
one_hot_encoded = pd.get_dummies(merged_df['article_id'], prefix='article')
one_hot_encoded = pd.concat([merged_df[['web_order_id']], one_hot_encoded], axis=1)

# Group by web_order_id to aggregate the one-hot encoded columns
one_hot_aggregated = one_hot_encoded.groupby('web_order_id').sum().reset_index()

# Aggregate the box_id column using the distinct count
box_id_count = merged_df.groupby('web_order_id')['box_id'].nunique().reset_index()
box_id_count.columns = ['web_order_id', 'unique_box_count']

# Aggregate the article_weight_in_g column using the sum
total_weight = merged_df.groupby('web_order_id')['article_weight_in_g'].sum().reset_index()
total_weight.columns = ['web_order_id', 'total_weight_in_g']
```

```python
# For the remaining columns, take the first value for each web_order_id
remaining_columns = merged_df.drop(columns=['article_id', 'box_id', 'article_weight_in_g'])\
                             .drop_duplicates(subset='web_order_id')

# Merge all aggregated dataframes on web_order_id
final_df = remaining_columns.merge(box_id_count, on='web_order_id') \
                            .merge(total_weight, on='web_order_id') \
                            .merge(one_hot_aggregated, on='web_order_id')

# Display the first few rows to verify
display(final_df.head())
```
