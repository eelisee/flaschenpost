```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df =pd.read_parquet(r"C:\Users\simon\OneDrive\Desktop\Datathon25\merged_df.parquet")
df
```

```python
df_filtered = df[df['service_time_in_minutes'] > 30]
df_filtered.sort_values(by='service_time_in_minutes', ascending=False).head(30)
```

```python
# Group by warehouse_id and count the number of orders
warehouse_order_counts = df_filtered.groupby('warehouse_id').size().reset_index(name='order_count')

# Display the result
warehouse_order_counts.sort_values('order_count', ascending=False)
```

```python
# Group by customer_id_x and count the number of orders
user_order_counts = df_filtered.groupby('customer_id_x').size().reset_index(name='order_count')

# Display the result
user_order_counts.sort_values('order_count', ascending=False)
```

```python
user_order_counts_filtered = user_order_counts[user_order_counts['order_count'] > 1]
user_order_counts_filtered
```

```python
total_orders = user_order_counts_filtered['order_count'].sum()
total_orders
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df_filtered['service_time_in_minutes'], shade=True)
plt.title('Density Plot of Service Time in Minutes')
plt.xlabel('Service Time in Minutes')
plt.ylabel('Density')
plt.show()
```
